```python
#wap in py accept a total unit number and per unit is 4 rs fixed and given cons
# unit range is
# 1 to 200 => No tax
# 200 to 300 => 2%
# 300 to 400 => 4%
# 400 to 500 => 6%
# 500 above => 10%
#

# Result print in this format

# #

# Total unit is :
# Total bill amount without tax is :
# Tax in percenatge is :
# Tax in rs is :
# Final bill payable amount is :

unit = int (input("Enter your Unit: "))

print ("=======================")
print ("Total unit is: ",unit)

#per unit is 4 rs fixed
total = (unit*4)
print ("Total bill amount without tax is : ",total)

# 1 to 200 => No tax
if unit>= 0 and unit<200:
    print("No tax")
    
# 200 to 300 => 2%
elif unit>= 200 and unit<300:
    print("Tax in percenatge is : 2%")
    tax=  0.02*total
    print("Tax in rs is :", tax)
    
# 300 to 400 => 4%
elif unit>= 300 and unit<400:
    print("Tax in percenatge is : 4%")
    tax=  0.04*total
    print("Tax in rs is :", tax)
    
# 400 to 500 => 6%
elif unit>= 400 and unit<500:
    print("Tax in percenatge is : 6%")
    tax=  0.06*total
    print("Tax in rs is :", tax)
    
# 500 above => 10%
elif unit>= 500:
    print("Tax in percenatge is : 10%")
    tax=  0.10*total
    print("Tax in rs is :", tax)

final_bill = total+tax
print("Final bill payable amount is :", final_bill)
print ("=======================")
```

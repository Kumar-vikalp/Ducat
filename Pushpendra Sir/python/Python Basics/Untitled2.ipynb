```python
def calculate_love(name1, name2):
    combined_names = name1 + name2
    love_score = 0

    for char in combined_names:
        love_score += ord(char)

    love_percentage = (love_score % 100) + 1  # Ensures the percentage is between 1 and 100
    return love_percentage

# Get names from user
name1 = input("Enter Your name: ")
name2 = input("Enter other name: ")

# Calculate love percentage
love_percentage = calculate_love(name1, name2)
print(f"The love percentage between {name1} and {name2} is: {love_percentage}%")
```

```python
import requests

def get_sun_sign(day, month):
    if (month == 3 and day >= 21) or (month == 4 and day <= 19):
        return "Aries"
    elif (month == 4 and day >= 20) or (month == 5 and day <= 20):
        return "Taurus"
    elif (month == 5 and day >= 21) or (month == 6 and day <= 20):
        return "Gemini"
    elif (month == 6 and day >= 21) or (month == 7 and day <= 22):
        return "Cancer"
    elif (month == 7 and day >= 23) or (month == 8 and day <= 22):
        return "Leo"
    elif (month == 8 and day >= 23) or (month == 9 and day <= 22):
        return "Virgo"
    elif (month == 9 and day >= 23) or (month == 10 and day <= 22):
        return "Libra"
    elif (month == 10 and day >= 23) or (month == 11 and day <= 21):
        return "Scorpio"
    elif (month == 11 and day >= 22) or (month == 12 and day <= 21):
        return "Sagittarius"
    elif (month == 12 and day >= 22) or (month == 1 and day <= 19):
        return "Capricorn"
    elif (month == 1 and day >= 20) or (month == 2 and day <= 18):
        return "Aquarius"
    elif (month == 2 and day >= 19) or (month == 3 and day <= 20):
        return "Pisces"
    else:
        return "Invalid date"

def get_relationship_advice(sign1, sign2):
    # Mock API URL
    api_url = f"https://mock-api.com/relationship-advice?sign1={sign1}&sign2={sign2}"
    
    response = requests.get(api_url)
    if response.status_code == 200:
        data = response.json()
        return data.get("advice", "No advice available.")
    else:
        return "Failed to fetch advice."

# Example usage
day = int(input("Enter your birth day: "))
month = int(input("Enter your birth month: "))
sun_sign1 = get_sun_sign(day, month)
print(f"Your sun sign is: {sun_sign1}")

day = int(input("Enter your partner's birth day: "))
month = int(input("Enter your partner's birth month: "))
sun_sign2 = get_sun_sign(day, month)
print(f"Your partner's sun sign is: {sun_sign2}")

advice = get_relationship_advice(sun_sign1, sun_sign2)
print(f"Relationship advice for {sun_sign1} and {sun_sign2}: {advice}")
```

```python
import requests
from bs4 import BeautifulSoup
import json

# URL of the page to scrape
url = "https://watch.plex.tv/movie/moana-2-2024-1"

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Function to convert BeautifulSoup elements to a dictionary
    def element_to_dict(element):
        if isinstance(element, str):
            return element.strip()
        if element.name:
            result = {"tag": element.name}
            if element.attrs:
                result["attributes"] = element.attrs
            if element.contents:
                children = [element_to_dict(child) for child in element.contents]
                children = [child for child in children if child]  # Remove empty children
                if children:
                    result["children"] = children
            return result
        return None
    
    # Convert the BeautifulSoup object to a dictionary
    json_data = element_to_dict(soup)
    
    # Print the JSON data
    print(json.dumps(json_data, indent=4))

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
